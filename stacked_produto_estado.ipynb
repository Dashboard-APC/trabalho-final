```python
# Importando os pacotes

import pandas as pd
import plotly.express as px
import numpy as np
#from google.colab import files
#import io
```

```python
# Importando e manipulando os dados

#uploaded = files.upload()  # Sobe arquivo do computador
#data = pd.read_csv(io.BytesIO(uploaded['2004-2021.tsv']), sep='\t') # Carrega arquivo recém subido

data = pd.read_csv('2004-2021.tsv', sep='\t') # Carregando arquivo upado manualmente no colab
data
```

```python
# Padronizando os dados para o gráfico

data['PRODUTO'] = data['PRODUTO'].replace(['OLEO DIESEL'],'ÓLEO DIESEL') # Unificando as duas ocorrências em uma só
data['PRODUTO'] = data['PRODUTO'].replace(['OLEO DIESEL S10'],'ÓLEO DIESEL S10') # Unificando as duas ocorrências em uma só

# Contando a quantidade de produtos com sum() que retorna os valores no formato de 'series' e transformando de volta para o formato dataframe com reset_index
data_produto = data.groupby(['ESTADO', 'PRODUTO'])['NÚMERO DE POSTOS PESQUISADOS'].sum().reset_index(name='NÚMERO DE POSTOS PESQUISADOS')
data_produto

# Contando os tipos de produtos por estado

```

```python
# Gráfico de barras empilhadas do tipo de produto por estado

bar_produto = px.bar(data_produto, x = 'ESTADO', y = 'NÚMERO DE POSTOS PESQUISADOS', color = 'PRODUTO')
bar_produto.update_layout(barmode='stack', xaxis={'categoryorder': 'total descending'}, 
                          yaxis_title = "Quantidade de postos", 
                          xaxis_title = 'Unidades Federativas') # Ordenando as barras e mudando o título do eixo x e y
bar_produto.show()
```
